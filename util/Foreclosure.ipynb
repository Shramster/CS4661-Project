```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt

def getRidOfDays(df):
    X = df
    X['Registered Date'] = df['Registered Date'].replace(to_replace="\/\d\d\/", regex=True, value="/")
    return X

# It's dd/mm/yyyy than mm/dd/yyyy for 2015
def getRidOfDays2015(df):
    X = df
    X['Registered Date'] = df['Registered Date'].replace(to_replace="^[0-9][0-9]\/", regex=True, value="")
    return X

# Remove the ".0" for 2018's Zip code
def removeDotZero(df):
    X = df
    return X

# Remove "12:00:00 AM" for 2021
def removeTime2021(df):
    X = df
    X['Registered Date'] = df['Registered Date'].replace(to_replace="\ \d\d\:\d\d\:\d\d\ [A-Z][A-Z]", regex=True, value="")
    return X
```

```python
df2014 = pd.read_csv("../source/2014_Registered_Foreclosure_Properties.csv")
df2015 = pd.read_csv("../source/2015_Registered_Foreclosure_Properties.csv")

df2016 = pd.read_csv("../source/2016_Registered_Foreclosure_Properties.csv").rename(columns= {"RegisteredDate":"Registered Date"})

df2017 = pd.read_csv("../source/2017_Registered_Foreclosure_Properties.csv")

df2018 = pd.read_csv("../source/2018_Registered_Foreclosure_Properties.csv")


df2019 = pd.read_csv("../source/2019_Registered_Foreclosure_Properties.csv")

df2020 = pd.read_csv("../source/2020_Registered_Foreclosure_Properties.csv")

df2021 = pd.read_csv("../source/2021_Registered_Foreclosure_Properties.csv")


df2022 = pd.read_csv("../source/2022_Registered_Foreclosure_Properties.csv")
df2014
```

```python
def remove_dups_and_non_residential(df):
    df = df[df['Property Type'] != 'Non-Residential']
    df = df[df.duplicated(subset=['Property Address']) != True]
    return df

def remove_dups_and_non_residential_2014(df):
    df = df[df['Property Type'] != 'Non-Residential']
    df = df[df.duplicated(subset=['Property Location']) != True]
    return df
df2014 = remove_dups_and_non_residential_2014(df2014)

df2015 = remove_dups_and_non_residential(df2015)

def remove_dups_and_non_residential_2016(df):
    df = df[df['PropertyType'] != 'Non-Residential']
    df = df[df.duplicated(subset=['PropertyAddress']) != True]
    return df
df2016 = remove_dups_and_non_residential_2016(df2016)

df2017 = remove_dups_and_non_residential(df2017)

df2018 = remove_dups_and_non_residential(df2018)

df2019 = remove_dups_and_non_residential(df2019)

df2020 = remove_dups_and_non_residential(df2020)

def remove_dups_and_non_residential_2021(df):
    df = df[df['Property Type'] != 'Non-Residential']
    df = df[df.duplicated(subset=['PropertyAddress']) != True]
    return df
df2021 = remove_dups_and_non_residential_2021(df2021)

df2022 = remove_dups_and_non_residential(df2022)
```

```python
df2014 = getRidOfDays(df2014)

df2015 = getRidOfDays2015(df2015)

df2016 = getRidOfDays(df2016)

df2017 = getRidOfDays(df2017)

df2018 = getRidOfDays(df2018)

df2019 = getRidOfDays(df2019)

df2020 = getRidOfDays(df2020)

df2021 = getRidOfDays(df2021)
df2021 = removeTime2021(df2021)

df2022 = getRidOfDays(df2022)
```
